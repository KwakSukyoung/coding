```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading Data

```python
df = pd.read_csv("piracydataset.csv")
df.set_index('Unnamed: 0', inplace=True)
```

```python
df
```

# Data Cleaning

```python
#1. remove exactly same rows
df = df.drop_duplicates()

df
```

```python
#2. check the multiple values of DISCORD_ID
# Get the counts of each value in the column
value_counts = df['discordid'].value_counts()

# Get values that appear multiple times
multiple_occurrences = value_counts[value_counts > 1].index.tolist()

# Filter the DataFrame based on values that occur multiple times in 'ColumnName'
rows_with_multiple_occurrences = df[df['discordid'].isin(multiple_occurrences)]

#Fix the unnecessary data
df = df.drop(184, axis=0) # same minecraftid but just capitalized
df = df.drop(483, axis=0) # same minecraftid but just capitalized
df = df.drop(517, axis=0) # just exactly the same row
df = df.drop(719, axis=0) # paid version but just different age
df.loc[88, 'age'] = 15.0 #change the age value since the age was 1
df.loc[294, 'age'] = 15.0 #change the age value since the age was 1
```
