```python
#(i) Two clusters by sampling 50 poins from a multivariate normal distribution with mena (0,0,0)
cluster1 = np.random.multivariate_normal((0,0,0), np.eye(3), size = 50)
cluster2 = np.random.multivariate_normal((3,3,3), np.eye(3), size = 50)
X = np.vstack((cluster1, cluster2))
y = np.concatenate((np.zeros(50), np.ones(50)))

perplexity = [2,5,10,30,50,100,500]

for i in perplexity:
    tsne = TSNE(n_components=2, perplexity = i).fit_transform(X)
    
    plt.scatter(tsne[:,0], tsne[:,1],c = y)
    plt.show()
```

```python
#(ii) Two clusters by sampling 50 poins from a multivariate normal distribution with mena (0,0,0)
cluster1 = np.random.multivariate_normal((0,0,0), np.eye(3), size = 50)
cluster2 = np.random.multivariate_normal((2,2,2), np.eye(3), size = 50)
cluster3 = np.random.multivariate_normal((-1,2,1), np.eye(3), size = 50)

X = np.vstack((cluster1, cluster2, cluster3))
y = np.concatenate((np.zeros(50), np.ones(50), 3*np.ones(50)))

perplexity = [2,5,10,30,50,100,500]

for i in perplexity:
    tsne = TSNE(n_components=2, perplexity = i).fit_transform(X)
    
    plt.scatter(tsne[:,0], tsne[:,1],c = y)
    plt.show()
```

t-SNE converts the similarities into probabilities.

UMAP converts similarities into lower-dimensional similarities.

t-SNE is better when the data has well-defined clusters.

```python

import sklearn.datasets
fmnist = sklearn.datasets.fetch_openml('Fashion-MNIST')
data, labels = fmnist.data, fmnist.target
```

```python
X = data[:4000]
y = labels[:4000]

tsne = TSNE(n_components=2).fit_transform(X)
colr = ['red','orange', 'yellow', 'green', 'blue','indigo','violet','pink','black','deeppink']
colors = [colr[int(i)] for i in list(y)]

plt.scatter(tsne[:,0], tsne[:,1],c = colors, s=1)
plt.show()

```
