```python
import numpy as np
from scipy.integrate import solve_ivp
from matplotlib import pyplot as plt
import math
```

```python
def secant_method(h, s0, s1, max_iter=100, tol=1e-8):
    """
    Finds a root of h(s)=0 using the secant method with the
    initial guesses s0, s1.
    """
    for i in range(max_iter):
        # Get the residuals
        h0 = h(s0)
        h1 = h(s1)
        # Update
        s2 = s1 - h1 * (s1 - s0)/(h1 - h0)
        s0, s1 = s1, s2
        
        # Check convergence
        if abs(h1) < tol:
            return s2
    
    print("Secant method did not converge")
    return s2
```

# Problem 1

Appropriately defined initial value problems will usually have a unique solution.
Boundary value problems are not so straightforward; they may have no solution or they may have several, and you may have to determine which solutions are physically interesting.

Use the secant method to solve the following BVP:
\begin{equation*}
\begin{split}
y'' &= -e^{y-1}, \quad x \in [0,1],\\
y(0) &=y(1) =1.
\end{split}
\end{equation*}
This BVP has two solutions.
Using the secant method, find both numerical solutions and their initial slopes.
What initial values $s_0, s_1$ did you use to find them?

```python
# Define the ODE right-hand side
def ode(x, y):
    return np.array([
        y[1],
        -np.exp(y[0]-1)
    ])
# Endpoint values
a =0
b = 1
alpha = 1
beta = 1
# Define a residual function
def residual(s):
    # Find the right endpoint
    sol = solve_ivp(ode, (a, b), [alpha, s])
    yb = sol.y[0,-1]
    return yb - beta
# Find the right value of s using the secant method
s0 = secant_method(residual, (beta-alpha)/2, -1)
# Compute and plot the solution
x = np.linspace(0,1,100)
y = solve_ivp(ode, (a,b), (alpha, s0), t_eval=x).y[0]
plt.plot(x, y)

# Find the right value of s using the secant method
s0 = secant_method(residual, (beta-alpha)/2, 9)
# Compute and plot the solution
x = np.linspace(0,1,100)
y = solve_ivp(ode, (a,b), (alpha, s0), t_eval=x).y[0]
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Both solutions to the boundary value problem given in Problem 1")

plt.show()
```

# Problem 2

Use Newton's method to solve the BVP
\begin{align*}
y'' &= 3 + \frac{2y}{x^2}, \,\, x \in [1,e],\\
y(1) &= 6, \\
y(e) &= e^2 + 6/e
\end{align*}
Plot your solution.

```python
# Define the ODE right-hand side
def ode(x, y):
    return np.array([
        y[1],
        3+(2*y[0]/(x**2)),
        y[3],
        y[2]*(2/(x**2))
    ])
# Endpoint values
a =1
b = math.e
alpha = 6
beta = ((math.e**2)+(6/math.e))
# Define a residual function
def residual(s):
    # Find the right endpoint
    sol = solve_ivp(ode, (a, b), [alpha, s, 0, 1])
    yb = sol.y[0,-1]
    zb = sol.y[2,-1]
    return (yb - beta)/zb

#Netons
def Neton(residual, init, max_iter = 100):
    for i in range(max_iter):
        init-=residual(init)
    return init
s0 = Neton(residual,3)
# Compute and plot the solution
x = np.linspace(a,b,100)
y = solve_ivp(ode, (a,b), (alpha, s0,0,1), t_eval=x).y[0]
plt.plot(x, y)
plt.title("SOlution for problem 2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Use the secant method to find initial values for $\theta$ that give solutions to the following BVP:
\begin{align}
	\begin{split}
\frac{dy}{dx} &= \tan {\theta} ,\\
\frac{dv}{dx} &= -\frac{g \sin{\theta} + \mu v^2}{v \cos{\theta}},\\
\frac{d\theta}{dx} &= -\frac{g}{v^2},\\
y(0)&= y(195) = 0,\\
v(0) &= 45 \text{ m/s}
	\end{split}
\end{align}
($g = 9.8067\text{ m/s}^2$.)

Find and plot at least two of them.
