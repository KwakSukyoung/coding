# Volume 4: Introduction to Initial and Boundary Value Problems

```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.integrate import solve_bvp
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

### Problem 1
Use `solve_ivp` to solve for $y$ in the equation $y'' - y = sin(t)$ with initial conditions $y(0)= -\frac{1}{2}$, $y'(0) = 0$ and plot your solution on the interval $[0,5]$. Compare this to the analytic solution $y=-\frac{1}{2}(e^{-t}+sin(t))$. 

Note: Using max_step = 0.1 with give you a smoother graph here

```python
# element 1: the ODE function
def ode(t, y):
    '''defines the ode system'''
    return np.array([y[1],np.sin(t)+y[0]])
# element 2: the time domain
t_span = (0,5)
# element 3: the initial conditions
y0 = np.array([-1/2,0])
# solve the system
# max_step is an optional parameter that controls maximum step size and
# a smaller value will result in a smoother graph
sol = solve_ivp(ode, t_span, y0, max_step=0.1)

#plotting solve_ivp
plt.subplot(2,1,1)
plt.plot(sol.t,sol.y[0])
plt.title("solve_ivp")
plt.xlabel('t')
plt.ylabel('y(t)')

#plotting actual solution
plt.subplot(2,1,2)
t = np.linspace(0,5,100)
y = -(np.exp(-t)+np.sin(t))/2
plt.plot(t,y, 'orange')
plt.title("actual solution")
plt.xlabel('t')
plt.ylabel('y(t)')

plt.tight_layout()
plt.show()
```

### Problem 2

Use `solve_bvp` to solve for $y$ in the equation $y'+y''=-\frac{1}{4}e^{-0.5t}+sin(t)-cos(t)$ with boundary conditions $y(0)=6$ and $y'(5)=-0.324705$ and plot your solution on the interval $[0,5]$. Use an initial guess of all ones. Compare this to the analytic solution $y =e^{-0.5t}-sin(t)+5$. 

```python
# element 1: the ODE function
def ode(t,y):
    ''' define the ode system '''
    return np.array([y[1], -np.exp(-t/2)/4 +np.sin(t)-np.cos(t)-y[1]])
# element 2: the boundary condition function
def bc(ya,yb):
    ''' define the boundary conditions '''
    # ya are the initial values
    # yb are the final values
    # each entry of the return array will be set to zero
    return np.array([ya[0] - 6, yb[1] + 0.324705])
# element 3: the time domain.
t_steps = 100
t = np.linspace(0,5,t_steps)
# element 4: the initial guess.
y0 = np.ones((2, t_steps))

# Solve the system.
sol = solve_bvp(ode, bc, t, y0)

#plotting solve_bvp
plt.subplot(2,1,1)
plt.plot(sol.x, sol.y[0])
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title("solve_bvp")
plt.tight_layout()

#plotting actual solution
plt.subplot(2,1,2)
t = np.linspace(0,5,100)
y = np.exp(-t/2)-np.sin(t)+5
plt.plot(t,y, 'orange')
plt.title("actual solution")
plt.xlabel('t')
plt.ylabel('y(t)')

plt.tight_layout()
plt.show()
```

### Problem 3

Use `solve_bvp` to solve for $y$ in the equation $y''=(1-y')*10y$ with boundary conditions $y(0)=-1$ and $y(1)=\frac{3}{2}$ and plot your solution on the interval $[0,1]$. Use an initial guess of 10. Compare this to the the same solution using an initial guess of 9.99. 

```python
# element 1: the ODE function
def ode(t,y):
    ''' define the ode system '''
    return np.array([y[1], (1-y[1])*10*y[0]])
# element 2: the boundary condition function
def bc(ya,yb):
    ''' define the boundary conditions '''
    # ya are the initial values
    # yb are the final values
    # each entry of the return array will be set to zero
    return np.array([ya[0] + 1, yb[0] - 3/2])
# element 3: the time domain.
t_steps = 50
# element 4: the initial guess.
t = np.linspace(0,1,t_steps)
y0 = np.ones((2, t_steps))
y0 *= 9.99

# Solve the system.
sol = solve_bvp(ode, bc, t, y0)

#plotting solve_bvp
plt.plot(sol.x, sol.y[0], color = 'blue',label = 'initial guess: 9.99')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.tight_layout()
plt.legend()

y0 = np.ones((2, t_steps))
y0 *= 10
# Solve the system.
sol = solve_bvp(ode, bc, t, y0)
#plotting solve_bvp
plt.plot(sol.x, sol.y[0], color = 'orange',label = 'initial guess: 10.00')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.tight_layout()
plt.legend()

plt.title('The differences of initial guesses 9.99 and 10.00')
# plt.tight_layout()
plt.show()
```

### Problem 4
Use `solve_ivp` to solve the Four-Wing Attractor as described in equations (1.2), (1.3), and (1.4) where $a=0.2$, $b = 0.01$, and $c = -0.4$. Try this with 3 different initial values and plot (in three dimensions) the 3 corresponding graphs.

```python
a = .2
b = .01
c = -.4

y = np.linspace(0,5)
z = np.linspace(0,5)

# element 1: the ODE function
def ode(t, s):
    '''defines the ode system'''
    dx = a*s[0]+s[1]*s[2]
    dy = b*s[0] +c*s[1] -s[0]*s[2]
    dz = -s[2]-s[0]*s[1]
    return np.array([dx,dy,dz])

# element 2: the time domain
t_span = (0,250)
# element 3: the initial conditions
y0 = np.array([2,2,2.5])


# solve the system
# max_step is an optional parameter that controls maximum step size and
# a smaller value will result in a smoother graph
sol = solve_ivp(ode, t_span, y0, max_step=0.1)
fig = plt.figure()

ax1 = fig.add_subplot(1,5,1, projection = '3d', title = "when y0 = 2,2,2.5")
ax1.plot(sol.y[0,:],
        sol.y[1,:],
        sol.y[2,:])
fig.set_figwidth(15)
fig.set_figheight(15)
plt.tight_layout()

# element 3: the initial conditions
y0 = np.array([2,2,3])
# solve the system
# max_step is an optional parameter that controls maximum step size and
# a smaller value will result in a smoother graph
sol = solve_ivp(ode, t_span, y0, max_step=0.1)

ax2 = fig.add_subplot(1,5,3, projection = '3d', title = "when y0 = 2,2,3")
ax2.plot(sol.y[0,:],
        sol.y[1,:],
        sol.y[2,:])
fig.set_figwidth(15)
fig.set_figheight(15)
plt.tight_layout()

# element 3: the initial conditions
y0 = np.array([1.5,2,2.5])
# solve the system
# max_step is an optional parameter that controls maximum step size and
# a smaller value will result in a smoother graph
sol = solve_ivp(ode, t_span, y0, max_step=0.1)

ax3 = fig.add_subplot(1,5,5, projection = '3d', title = "when y0 = 1.5,2,2.5")
ax3.plot(sol.y[0,:],
        sol.y[1,:],
        sol.y[2,:])
fig.set_figwidth(15)
fig.set_figheight(15)
plt.tight_layout()
plt.show()
```
