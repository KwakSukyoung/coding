## Problem 1
Write a function which implements Euler's method. Test your function on the IVP:
$$\begin{align}
	\begin{split}
		x' (t)&= x(t) - 2t + 4,\quad 0 \leq t \leq 2, \\
		x(0) &= 0,
	\end{split}
\end{align}$$

where the analytic solution is $x(t) = -2+2t + 2e^t$.

Use the Euler method to numerically approximate the solution with step sizes $h = 0.2, 0.1$, and $0.05.$  Plot the true solution alongside the three approximations.

```python
def euler(f,x0,t):
    """Numerically approximates the solution to the IVP:
    
    x'(t) = f(x(t),t)
    x(t0) = x0
    
    using the Euler method.
    Parameters:
        f (function): The right-hand side of the ODE
        x0 ((m,) ndarray): The initial condition
        t ((n,) ndarray): The array of time values
    Returns:
        ((n,m) ndarray): The approximate solution, where x[i] â‰ˆ x(t_i)
    """
    #setting x
    n = len(t)
    m = len(x0)
    h = t[1]-t[0]
    x = np.zeros((n, m))
    
    #changing x
    for i in range(n-1):
        x[i+1] = x[i] + h*f(x[i], t[i])
        
    return x
```

```python
#setting f,x0, and y
f = lambda x, t: x - (2*t) + 4
x_0 =[0]
y = lambda t: -2+(2*t)+2*np.exp(t)

#plotting for each h
for h in [0.2, 0.1, 0.05]: 
    t = np.arange(0, 2+h, h)
    plt.plot(t, euler(f, x_0, t), label = "h = " + str(h))

#plotting
plt.plot(t, y(t), '--', label = "Analytical solution")
plt.xlabel("t")
plt.ylabel("x")
plt.title("Alongside several approximations using Euler's Method")
plt.legend()

plt.show() 
```

```python
def midpoint(f,x0,t):
    """Numerically approximates the solution to the IVP:
    
    x'(t) = f(x(t),t)
    x(t0) = x0
    
    using the midpoint method.
    Parameters:
        f (function): The right-hand side of the ODE
        x0 ((m,) ndarray): The initial condition
        t ((n,) ndarray): The array of time values
    Returns:
        ((n,m) ndarray): The approximate solution, where x[i] â‰ˆ x(t_i)
    """
    #setting x
    n = len(t)
    m = len(x0)
    h = t[1]-t[0]
    x = np.zeros((n, m))
    
    #changing x
    for i in range(n-1):
        x[i+1] = x[i] + h*f(x[i] + h*(f(x[i], t[i]))/2, t[i] + (h/2))

    return x

def rk4(f,x0,t):
    """Numerically approximates the solution to the IVP:
    
    x'(t) = f(x(t),t)
    x(t0) = x0
    
    using a fourth-order Runge-Kutta method.
    Parameters:
        f (function): The right-hand side of the ODE
        x0 ((m,) ndarray): The initial condition
        t ((n,) ndarray): The array of time values
    Returns:
        ((n,m) ndarray): The approximate solution, where x[i] â‰ˆ x(t_i)
    """
    #setting x
    n = len(t)
    m = len(x0)
    h = t[1]-t[0]
    x = np.zeros((n, m))
    x[0]= x0
    
    #changing x
    for i in range(n-1):
        K1 = f(x[i], t[i])                   
        K2 = f(x[i] + h*K1/2, t[i] + (h/2))
        K3 = f(x[i] + h*K2/2, t[i] + (h/2))
        K4 = f(x[i] + h*K3, t[i+1])          

        x[i+1] = x[i] + h*(K1 + 2*K2 + 2*K3 + K4)/6

    return x
```

## Problem 3

Use the RK4 method to solve for the simple harmonic oscillator satisfying:
$$\begin{align}
	\begin{split}
&{}my'' + ky = 0,\quad 0 \leq t \leq 20, \\
&{}y(0) = 2, \quad
y'(0) = -1,
	\end{split}
\end{align}$$

```python
#setting the variables
m1, m2, k = 1, 3, 1
x1 = lambda x,t: np.array([x[1], -k*x[0]/m1])
x2 = lambda x,t: np.array([x[1], -k*x[0]/m2])
x_0 =[2,-1]
t = np.arange(0, 20.05, .05)

#plotting
plt.plot(t, rk4(x1,x_0,t)[:,0], color = 'black',label = "m = 3")
plt.plot(t, rk4(x2,x_0,t)[:,0], color = 'blue', label = "m = 1")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Harmonic Oscillator")
plt.legend()

plt.show()
```

## Problem 4

Use the RK4 method to solve for the damped free harmonic oscillator satisfying 
$$\begin{align*}
&{}y'' +\gamma y'+ y = 0, \quad 0 \leq t \leq 20,\\
&{}y(0) = 1, \quad
y'(0) = -1.
\end{align*}$$
For $\gamma = 1/2,$ and $\gamma = 1$, simultaneously plot your numerical approximations of $y$.

```python
#setting the variables
m, r1, r2, k = 1, 1/2, 1, 1
x1 = lambda x,t: np.array([x[1], -k*x[0] - r1*x[1]/m])
x2 = lambda x,t: np.array([x[1], -k*x[0] - r2*x[1]/m])
x_0 =[1,-1]
t = np.arange(0, 20.05, .05)

#plotting
plt.plot(t, rk4(x1,x_0,t)[:,0], color = 'black',label = r"$\gamma$ = 1/2")
plt.plot(t, rk4(x2,x_0,t)[:,0], color = 'blue', label = r"$\gamma$ = 1")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Damped Free Simple Harmonic Oscillator")
plt.legend()

plt.show()
```

## Problem 5

Use the RK4 method to solve for the damped and forced harmonic oscillator satisfying 
\begin{align}
	\begin{split}
&{}2y'' + \gamma y' + 2y = 2 \cos (\omega t), \quad 0 \leq t \leq 40,\\
&{}y(0) = 2, \quad
y'(0) = -1. 
	\end{split}
	\label{ivp:damped_forced_oscillator}
\end{align}
For the following values of $\gamma$ and $\omega,$ plot your numerical approximations of $y(t)$: $(\gamma, \omega) = (0.5, 1.5),$ $(0.1, 1.1),$ and $(0, 1)$.

```python
#setting the variables
m, k = 2, 2
r1, r2, r3 = .5, .1, 0
w1, w2, w3 = 1.5, 1.1, 1
x1 = lambda x,t: np.array([x[1], (-k*x[0] - r1*x[1])/m + 2*np.cos(w1*t)])
x2 = lambda x,t: np.array([x[1], (-k*x[0] - r2*x[1])/m + 2*np.cos(w2*t)])
x3 = lambda x,t: np.array([x[1], (-k*x[0] - r3*x[1])/m + 2*np.cos(w3*t)])
x_0 =[1,-1]
t = np.arange(0, 40.05, .05)

#plotting
plt.plot(t, rk4(x1,x_0,t)[:,0], color = 'black',label = r"$\gamma$ = .5, $\omega$ = 1.5")
plt.plot(t, rk4(x2,x_0,t)[:,0], color = 'blue', label = r"$\gamma$ = 1, $\omega$ = 1.1")
plt.plot(t, rk4(x3,x_0,t)[:,0], color = 'green', label = r"$\gamma$ = 1, $\omega$ = 1")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Damped and Forced Harmonic Oscillator")
plt.legend()

plt.show()
```
