```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton
```

# Problem 1

Use the natural embedding algorithm to create a bifurcation diagram for the differential equation

x = λx − x^3

This type of bifurcation is called a pitchfork bifurcation (you should see a pitchfork in your
diagram)

```python
def EmbeddingAlg(param_list, guess, F):
    X = []
    for param in param_list:
        try:
            # Solve for x_value making F(x_value, param) = 0.
            x_value = newton(F, guess, fprime=None, args=(param,), tol=1E-7, maxiter=50)
            # Record the solution and update guess for the next iteration.
            X.append(x_value)
            guess = x_value
        except RuntimeError:
            # If Newton's method fails, return a truncated list of parameters
            # with the corresponding x values.
            return param_list[:len(X)], X
    # Return the list of parameters and the corresponding x values.
    return param_list, X
```

```python
def F(x, lmbda):
    return -x**3 + lmbda*x

# Top curve shown in the bifurcation diagram
C1, X1 = EmbeddingAlg(np.linspace(5, 0, 200), np.sqrt(5), F)
# The bottom curve
C2, X2 = EmbeddingAlg(np.linspace(5, 0, 200), -np.sqrt(5), F)
# The horizon
C3, X3 = EmbeddingAlg(np.linspace(5, -5, 200), -np.sqrt(0), F)
plt.title("x = λx − x^3")

plt.plot(C1,X1)
plt.plot(C2,X2)
plt.plot(C3,X3)
```

# Problem 3 - Hysteresis

Reproduce the bifurcation diagram for the differential equation

dx/dτ = rx(1 − x/k) −x^2/(1 + x)^2,

where r = 0.56

```python
def EmbeddingAlg(param_list, guess, F):
    X = []
    for param in param_list:
        try:
            # Solve for x_value making F(x_value, param) = 0.
            x_value = newton(F, guess, fprime=None, args=(param,), tol=1E-7, maxiter=50)
            # Record the solution and update guess for the next iteration.
            X.append(x_value)
            guess = x_value
        except RuntimeError:
            # If Newton's method fails, return a truncated list of parameters
            # with the corresponding x values.
            return param_list[:len(X)], X
    # Return the list of parameters and the corresponding x values.
    return param_list, X
```

```python
r = .56
def F(x, k):
    return r*x*(1-x/k)-((x**2)/(1+x**2))

# Top curve shown in the bifurcation diagram
C1, X1 = EmbeddingAlg(np.linspace(2, 10, 200), np.sqrt(5), F)
# The bottom curve
C2, X2 = EmbeddingAlg(np.linspace(14, 5, 200), 7, F)
# The horizon
C3, X3 = EmbeddingAlg(np.linspace(8, 10, 200), 3, F)

C4, X4 = EmbeddingAlg(np.linspace(6.6, 8, 200), 2.2, F)

C5, X5 = EmbeddingAlg(np.linspace(6.6, 6, 200), 2.2, F)
plt.title("Budworm Population, Bryce said that we don't need the dots")

plt.plot(C1,X1, color = 'black', label = 'Stable Equilibria')
plt.plot(C2,X2, color = 'black')
plt.plot(C3,X3, color = 'black', linestyle = 'dashed', label = 'Unstable Equilibria')
plt.plot(C4,X4, color = 'black', linestyle = 'dashed')
plt.plot(C5,X5, color = 'black', linestyle = 'dashed')
plt.xlabel('k')
plt.legend()
plt.show()
```
