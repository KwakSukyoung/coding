```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

# Problem 1

Solve (16.2) for a(x).
Produce the plot shown in Figure 16.2.

```python
#given information
c0= 3/8
c1 = 5/4
X = np.array([.1*j for j in range(11)])
u = lambda x: (x**2)+(x/2)+(5/16)
f = lambda x: (-6*(x)**2)+(3*x)-1 if x<=1/2 else -1

#get as
a = []
for x in X:
    top = c0-integrate.quad(f, 0, x)[0]
    bottom = (2*x)+(1/2)
    a.append(top/bottom)

#plotting
plt.plot(a)
plt.title("a(x) for problem1")
plt.show()
```

# Problem 2

Find the density function a(x). Using several values of Îµ > 0.66049142, plot the corresponding
density a(x) for x in np.linspace(0,1,11) to demonstrate that the problem is ill-posed.

```python
#given information
c0 = 1
c1 = 2
eps = .8
X = np.linspace(0,1,11)
u = lambda x: x+1+eps*(np.sin(x*(eps**-2)))
f = lambda x: -1

#get as
a = []
for x in X:
    top = c0-integrate.quad(f, 0, x)[0]
    bottom = 1-(eps*-1)*(np.cos(x*(eps**-2)))
    a.append(top/bottom)

plt.plot(a,label= r"when $\epsilon$=.8")

#given information
c0 = 1
c1 = 2
eps = .7
X = np.linspace(0,1,11)
u = lambda x: x+1+eps*(np.sin(x*(eps**-2)))
f = lambda x: -1

#get as
a = []
for x in X:
    top = c0-integrate.quad(f, 0, x)[0]
    bottom = 1-(eps*-1)*(np.cos(x*(eps**-2)))
    a.append(top/bottom)

#plotting
plt.plot(a,label= r"when $\epsilon$=.7")


#given information
c0 = 1
c1 = 2
eps = .6
X = np.linspace(0,1,11)
u = lambda x: x+1+eps*(np.sin(x*(eps**-2)))
f = lambda x: -1

#get as
a = []
for x in X:
    top = c0-integrate.quad(f, 0, x)[0]
    bottom = 1-(eps*-1)*(np.cos(x*(eps**-2)))
    a.append(top/bottom)

plt.plot(a,label= r"when $\epsilon$=.6")
plt.title("a(x) for problem2")
plt.legend()
plt.show()
```
