# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    #read the data
    data = pd.read_csv("crime_data.csv", index_col = "Year")
    #1st visualization
    data.plot(y="Total", title = "Total", ylabel = "Total")
    #2nd visualization
    data.plot.scatter(x= "Total", y = "Population", title = "Total vs Population")
    #3rd visulization
    data.plot(kind = 'hist', y = ['Property'], bins = 10, stacked = True, title = "Frequencey of Simailar Crimes")
    plt.xlabel("Degree")
    plt.legend()
```

```python
prob1()
```

# Problem 2

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    #read the data 
    data = pd.read_csv("crime_data.csv", index_col = "Year")
    #trend between Forcible Rape and Violent
    pd.plotting.scatter_matrix(data[['Forcible Rape', 'Violent']])
    plt.suptitle("Violent vs Forcible Rape")
    #trend between Forcible Rape and Burglary
    pd.plotting.scatter_matrix(data[['Forcible Rape', 'Burglary']])
    plt.suptitle("Burglary vs Forcible Rape")
    #trend between Forcible Rape and Aggravated Assualt
    pd.plotting.scatter_matrix(data[['Forcible Rape', 'Aggravated Assault']])
    plt.suptitle("Aggravated Assault vs Forcible Rape")
    
    return (True, False, True)
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    data = pd.read_csv("crime_data.csv")
    #1st Plotting
    data.plot(kind = "box", y = ["Burglary", "Violent", "Vehicle Theft"], ylabel = "degree")
    plt.suptitle("The distributions of Burglary, Violent, and Vecicle Theft")
    #2nd Plotting
    data.plot(kind = "hexbin", x = "Vehicle Theft", y = "Robbery", gridsize = 20, sharex = False)
    plt.suptitle("Vehicle Theft vs Robbery")
    plt.tight_layout()
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    data = pd.read_csv("college.csv")
    #distribution of graduate rate
    data[['Grad.Rate']].hist(grid = False, bins = 10,sharex = False)
    plt.title('Distribution of Graduate Rate')
    plt.ylabel('Number of colleges')
    plt.xlabel('Graduation Rate')
    plt.text(0, -50, r"Source: James, et.al (2021) Data for an Introduction to Statistical Learning with Applications in R", fontsize = 10)
    plt.tight_layout()
    #Do schools with lower student to faculty ratios have higher tutition costs
    data.plot(kind='scatter', x = 'S.F.Ratio', y = 'Expend')
    plt.title('Lower student to Faculty ratios vs higher tution')
    plt.text(0, -15000, r"Source: James, et.al (2021) Data for an Introduction to Statistical Learning with Applications in R", fontsize = 10)
    plt.tight_layout()
    #Acceptance vs Enrollment
    data.plot(kind="box", y=["Accept","Enroll"])
    plt.title("Acceptance vs Enrollment")
    plt.ylabel("Frequency")
    plt.text(0, -10000, r"Source: James, et.al (2021) Data for an Introduction to Statistical Learning with Applications in R", fontsize = 10)
```
