```python
"""Volume 2: Non-negative Matrix Factorization."""

import numpy as np
import cvxpy as cp
from matplotlib import pyplot as plt
import os
from imageio import imread
import warnings
warnings.filterwarnings("ignore")
import random
from numpy import linalg as LA

from sklearn.decomposition import NMF
from sklearn.metrics import mean_squared_error as mse
from sklearn.datasets import load_sample_images
```

```python
#Problems 1-2
class NMFRecommender:

    def __init__(self,random_state=15,rank=3,maxiter=200,tol=1e-3):
        """The parameter values for the algorithm"""
        #setting things initialized
        self.random_state = random_state
        self.rank = rank
        self.maxiter = maxiter
        self.tol = tol
    
    def initialize_matrices(self,m,n):
        """randomly initialize the W and H matrices,"""
        #fix the random number
        np.random.seed(self.random_state)
        #get the rank number
        k = self.rank
        self.m = m
        self.n = n
        #get W and H using random.rand()
        self.W = np.random.rand(m,k)
        self.H = np.random.rand(k,n) 
        return self.W, self.H
      
    def fit(self,V):
        """Fits W and H weight matrices using CVXPY"""
        #loop through 0,....n-1
        for i in range(self.maxiter):
            #set W and H
            W = cp.Variable((self.rank, self.n), nonneg=True)
            H = cp.Variable((self.m, self.rank), nonneg=True)
            #getting self.H
            objective =  cp.Minimize(cp.norm(V - self.W @ W, 'fro'))
            problem = cp.Problem(objective)
            val = problem.solve()
            self.H = W.value 
            #getting self.W
            objective =  cp.Minimize(cp.norm(V - H @ self.H, 'fro'))
            problem = cp.Problem(objective)
            val = problem.solve()
            self.W = H.value 
            #when to stop
            if LA.norm(V - self.W @ self.H, ord='fro') < self.tol:
                break
    
    def reconstruct(self):
        """Reconstruct V matrix for comparison against the original V"""

        return self.W@ self.H
```

```python
def prob3():
    """Run NMF recommender on the grocery store example"""
    V = np.array(
        [
            [0, 1, 0, 1, 2, 2],
            [2, 3, 1, 1 ,2, 2],
            [1, 1, 1, 0, 1, 1],
            [0, 2, 3, 4, 1, 1],
            [0, 0, 0, 0, 1, 0]
        ]
    )
    #setting the matrix according to V
    Momo = NMFRecommender(rank=2)
    Momo.initialize_matrices(V.shape[0], V.shape[1])
    Momo.fit(V)
    #get W and H
    W = Momo.W
    H = Momo.H
    #count how much import it is
    element = sum([1 if np.argmax(H[:,i]) == 1 else 0 for i in range(H.shape[1])])
    
    return np.round(W,2), np.round(H,2), element
```

```python
def get_faces(path="./faces94"):
    """Traverse the specified directory to obtain one image per subdirectory.
        Flatten and convert each image to grayscale.
        Parameters:
            path (str): The directory containing the dataset of images.
Returns:
￿￿￿
((mn,k) ndarray) An array containing one column vector per
                subdirectory. k is the number of people, and each original
                image is mxn.
"""
    # Traverse the directory and get one image per subdirectory.
    faces = []
    for (dirpath, dirnames, filenames) in os.walk(path):
        for fname in filenames:
            if fname[-3:]=="jpg":       # Only get jpg images.
                # Load the image, convert it to grayscale,
                # and flatten it into a vector.
                faces.append(np.ravel(imread(dirpath+"/"+fname, as_gray=True)))
                break
    # Put all the face vectors column-wise into a matrix.
    return np.transpose(faces)
def show(image, m=200, n=180, plt_show=False,title=""):
    """Plot the flattened grayscale 'image' of width 'w' and height 'h'.
    Parameters:
        image ((mn,) ndarray): A flattened image.
        m (int): The original number of rows in the image.
        n (int): The original number of columns in the image.
        plt_show (bool): if True, call plt.show() at the end
"""
#scale image
    image = image / 255
    #reshape image
    image = np.reshape(image,(m,n))
#show image
    plt.imshow(image,cmap = "gray")
    plt.title = title
    if plt_show:
        plt.show()
```

```python
def prob4():
    """
        Gridsearch over rank, alpha and l1_ratio values to reconstruct 
        image using NMF. Plot all reconstructed images.
    """
    #loading the facial dataset
    image = get_faces()
    facial_data = image.T
    #reconstruct the third face
    show(facial_data[2], plt_show=True,title = "Original")
    
    #greed search
    for i in [0, 0.2, 0.5]:
        for j in [0, 10**(-5), 10]:
            #geting the componenets
            Christian = NMF(n_components=75, alpha=i, l1_ratio=j)
            W = Christian.fit_transform(image)
            H = Christian.components_ 
            
            show((W @ H).T[2], plt_show=True, title = "alpha: " + str(i) + ", l1_ratio: " + str(j))

```

```python
prob4()
```

```python
def prob5():
    '''
        find the 10 basis faces with the largest coefficients 
        corresponding to the the second and twelfth face in the dataset. 
        Plot these basis faces along with the original image using 
        subplots
    '''
    
    #loading the facial dataset
    dataset = get_faces()
    
    #2 and 12
    ind = [1,11]
    
    #second and twelveth faces
    show(dataset[:,ind[0]],plt_show=True, title = "The second Original")
    show(dataset[:,ind[1]],plt_show=True, title = "The 12th Original")
    
    #getting W and V
    Sam = NMF(n_components=75, alpha = .5,l1_ratio=0)
    W = Sam.fit_transform(dataset)
    H = Sam.components_
    
    #greed search
    for i in ind:
        for j in np.argsort(H[:,i])[-10:]:
            show(W[:,i], plt_show=True,title = str(j) + "nd iteration. ", str(i) + "th face")
```
