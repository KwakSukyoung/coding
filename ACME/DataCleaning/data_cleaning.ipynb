### Problem 3

basketball.csv contains data for all NBA players between 2001 and 2018.
Each row represents a player's stats for a year.

Create two new features:

    career_length (int): number of years player has been playing (start at 0).
    
    target (str): The target team if the player is leaving. If the player is retiring, the target should be 'retires'.
                  A player is retiring if their name doesn't exist the next year.
                  (Set the players in 2019 to NaN).

Remove all duplicate players in each year.
Remove all rows except those where a player changes team, that is, target is not null nor 'retires'.

Drop the player, year, and team_id columns.

Return the first 10 lines of your dataframe.

```python
basketball = pd.read_csv('basketball.csv')

#Creating two new features:[career_length]
basketball['career length'] = basketball.apply(lambda x: x['year'] - basketball.loc[basketball['player'] == x['player'],['year']].min(), axis = 1)

#Creating two new features:[target]
basketball.drop_duplicates(subset=['player', 'year'], inplace = True)
n_year = [basketball[basketball['player'] == player][basketball['year']==year+1]['team_id']for player, year in zip(basketball['player'], basketball['year'])]
basketball['target'] = ['retires' if len(item) == 0 else np.unique(item)[0] for item in n_year]

#Set the players in 2019 to NAN
basketball.loc[basketball['year'] == 2019, 'target'] == np.nan
basketball = basketball[basketball['target'] != basketball['team_id']]

#Remove all the rows except those where a player changes team, that is, target is not null nor 'retires'
basketball = basketball[basketball['year'] != 2019]
basketball = basketball[basketball['target'] != 'retires']

#Dropping the player, year, and team_id columns
basketball.drop(['player', 'year', 'team_id'], axis = 1, inplace = True)
print(basketball.head(10))
```

Load housing.csv into a dataframe with index=0. Descriptions of the features are in housing_data_description.txt.  
The goal is to construct a regression model that predicts SalePrice using the other features of the dataset.  Do this as follows:

	1) Identify and handle the missing data.  Hint: Dropping every row with some missing data is not a good choice because it gives you an empty dataframe.  What can you do instead?

	2) Identify the variable with nonnumeric values that are misencoded as numbers.  One-hot encode it. Hint: don't forget to remove one of the encoded columns to prevent collinearity with the constant column (which you will add later).
    
    3) Add a constant column to the dataframe.

    4) Save a copy of the dataframe.

	5) Choose four categorical featrues that seem very important in predicting SalePrice. One-hot encode these features and remove all other categorical features.
		
	6) Run an OLS using all numerical data regression on your model.  

	
Print the ten features that have the highest coef in your model and the summary. Don't print the OLS

```python
#load the data with index  = 0
housing = pd.read_csv('housing.csv',index_col = 0)

#1. Clean Data: 
#identify missing data 
obj_columns = housing.columns[(housing.isna().sum() != 0) & (housing.dtypes == object)]
num_columns = housing.columns[(housing.isna().sum() != 0) & (housing.dtypes != object)]

#handling the missing data for objective columns
for col in obj_columns:
    housing[col].fillna('N/A', inplace = True)

#handling the missing data for numeric columns
for col in num_columns:
    ran = list(np.linspace(housing[col].min(), housing[col].max(), housing[col].isna().sum()))
    for i, value in enumerate(ran):
        housing[col].fillna(value, limit=1, inplace=True)
housing.isna().sum()
```

```python
#Apply one-hot encoding
#Identify the variable with nonnumeric values that are misencoded as numbers
numeric_columns = [col for col in housing.columns if housing[col].dtypes != object]
housing = pd.get_dummies(housing, columns=['MSSubClass'], drop_first = True)

#Add a constant column to the dataframe.
housing['constant'] = 1

#Save a copy of the dataframe.
save_housing = housing.copy()

#Q.5:
# Choose four categorical featrues that seem very important in predicting SalePrice. 
# One-hot encode these features
housing = pd.get_dummies(housing, columns = ['HeatingQC', 'HouseStyle', 'PoolQC', 'SaleType'], drop_first = True)

# remove all other categorical features.
cols = housing.columns[housing.dtypes == 'object']
housing.drop(columns = cols, inplace = True)
print(housing.columns)
```

```python
#Run an OLS using all numerical data regression on your model.  
results = sm.OLS(housing['SalePrice'], housing.drop(columns = 'SalePrice')).fit()

# Print the summary
summ = results.summary()

# Convert the summary table to a dataframe
results_as_html = summ.tables[1].as_html()
result_df = pd.read_html(results_as_html, header=0, index_col=0)[0]

#Print the ten features that have the highest coefficient in your model and the summary.
result_df.sort_values(by = 'coef', ascending = False, inplace = True)
print(list(result_df[:10].index))
print(summ)
```
